```python
import pandas as pd
import os
import folium
from folium import IFrame
import base64


df = pd.read_csv("map_points.csv")
df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)

print(df)
```

```python
# Create two lists for the loop results to be placed
latlist = []
lonlist = []

# For each row in a varible,
for row in df['Geo']:
    # Try to,
    try:
        # Split the row by comma and append
        # everything before the comma to lat
        latlist.append(row.split(',')[0])
        # Split the row by comma and append
        # everything after the comma to lon
        lonlist.append(row.split(',')[1])
    # But if you get an error
    except:
        # append a missing value to lat
        latlist.append(np.NaN)
        # append a missing value to lon
        lonlist.append(np.NaN)

# Create two new columns from lat and lon
df['Lat'] = latlist
df['Lon'] = lonlist

print(df)
```

```python
#Create map object:
m=folium.Map(location=[51.0, 5.8],zoom_start=10)

#Create iFrame tree:
for Lat,Lon,Filepic,Filename,Type in zip(df['Lat'],df['Lon'],df['Filepic'],df['Filename'],df['Type']):
    encoded = base64.b64encode(open(Filepic, 'rb').read())
    html = '<img src="data:image/jpeg;base64,{}">'.format
    Filename = '<h3>{}</h3>'.format(Filename)
    iframe = IFrame(Filename + html(encoded.decode('UTF-8')), width=220, height=260)
    popup = folium.Popup(iframe, max_width=1000)
    if Type == 'tree':
        icon = folium.Icon(color="green", icon="tree-conifer")
    elif Type == 'food':
        icon = folium.Icon(color="red", icon="cutlery")
    elif Type == 'city':
        icon = folium.Icon(color="blue", icon="tower")    
    marker = folium.Marker(location=[Lat, Lon], popup=popup, icon=icon)
    marker.add_to(m)

m
```
